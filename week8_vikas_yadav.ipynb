# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus

````python
# 🛠️ TOOL 1: Calculator

def calculator(expression: str) -> str:
    """Evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# 🛠️ TOOL 2: Keyword Extractor

def extract_keywords(text: str) -> list:
    """Extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

## 🤖 Implement Agent Logic Below

👉 Use conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# 🤖 AGENT FUNCTION (TO IMPLEMENT)

import logging

# --- Bonus: basic logging setup ---
logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")
logger = logging.getLogger("agent")


def agent(query: str):
    """
    Single-agent router.

    Routing rules:
      - "calculate" in query -> Calculator Tool
      - "keywords"  in query -> Keyword Extractor Tool
      - otherwise            -> general response

    Always returns a dict of the form:
        {"type": "calculation" | "keywords" | "general" | "error", "result": ...}
    """
    # Basic input validation / error handling
    if not isinstance(query, str) or not query.strip():
        logger.warning("Empty or invalid query received.")
        return {"type": "error", "result": "Query must be a non-empty string."}

    query_lower = query.lower()
    logger.info(f"Received query: {query!r}")

    try:
        # --- Route 1: Math / Calculator ---
        if "calculate" in query_lower:
            logger.info("Routing to Calculator Tool.")

            # Pull out the mathematical expression from the query
            # (strip the trigger word and any non math-safe characters)
            expression = query_lower.replace("calculate", "")
            allowed_chars = set("0123456789+-*/(). ")
            expression = "".join(ch for ch in expression if ch in allowed_chars).strip()

            if not expression:
                return {"type": "error", "result": "No valid mathematical expression found in query."}

            calc_result = calculator(expression)

            if calc_result == "Error in calculation":
                return {"type": "error", "result": calc_result}

            return {"type": "calculation", "result": calc_result}

        # --- Route 2: Keyword extraction ---
        elif "keywords" in query_lower:
            logger.info("Routing to Keyword Extractor Tool.")

            # Remove the trigger phrase so it isn't extracted as a keyword itself
            text = query
            for phrase in ["extract keywords from", "keywords from", "extract keywords", "keywords"]:
                text = text.lower().replace(phrase, "") if phrase in text.lower() else text
            text = text.strip() or query

            kw_result = extract_keywords(text)

            if not kw_result:
                return {"type": "error", "result": "No keywords could be extracted from query."}

            return {"type": "keywords", "result": kw_result}

        # --- Route 3: General / fallback response ---
        else:
            logger.info("Routing to general response.")
            return {
                "type": "general",
                "result": f"I received your query: '{query}'. I currently only have tools for "
                           f"calculations (say 'calculate ...') and keyword extraction "
                           f"(say 'extract keywords from ...')."
            }

    except Exception as e:
        logger.error(f"Unexpected error while handling query: {e}")
        return {"type": "error", "result": f"Unexpected error: {e}"}
````

  "result": ...
}
